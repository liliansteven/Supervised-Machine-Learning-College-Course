Plot the points and lines

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8,10,12])
y = np.array([5,6,6,8,5,12])
y1 = x
y2 = 0.5*x
y3 = (0.5*x)+4
```

```python
plt.scatter(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

Calcutale the cost in eash case 

```python
def MSE(y_true,y_pred):
    error = (1/y_true.size)*(np.sum(np.square(y_true-y_pred)))
    return error
```

```python
error1 = MSE(y,y1)
error2 = MSE(y,y2)
error3 = MSE(y,y3)

print("Error1:",error1)
print("Error2:",error2)
print("Error3:",error3)
```

3d scatter to plot w , b , j(w,b)

```python
j1 = np.array([error1,error2,error3])
w1 = np.array([1,0.5,0.5])
b1 = np.array([0,0,4])

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(j1, w1, b1, color = "green")
plt.title("simple 3D scatter plot")
plt.show()
```
