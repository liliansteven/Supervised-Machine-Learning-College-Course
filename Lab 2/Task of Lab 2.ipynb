```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
 
x1 = np.zeros((67, 3)) 
x1[:, 0]=np.arange(-1,1,0.03) 
 
print(x1.shape) 
y1=x1[:, 0]**3+x1[:, 0]**2 
print(x1[:5]) 
print(y1[:5]) 
plt.scatter(x1[:, 0],y1) 
x1[:, 1]=x1[:, 0]**2 
x1[:, 2]=x1[:, 0]**3 
print(x1.shape) 
print(y1.shape)
```

```python
class Linear_Regression():
    def __init__(self, learning_rate , no_of_iterations ):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations

    def fit(self, X, Y):
        self.m  , self.n = X.shape  ## number of rows


        self.w=np.zeros(self.n)

        self.b = 0

        self.X = X

        self.Y = Y

        for i in range(self.no_of_iterations):
            self.update_weights()
            
        return self.w,self.b


    def update_weights(self):
        Y_prediction = self.predict(self.X)

        dw =  - ( ( (self.X.T).dot(self.Y - Y_prediction)) / self.m )+ self.w

        db =  - np.sum(self.Y - Y_prediction) / self.m

        self.w = self.w - self.learning_rate * dw

        self.b = self.b - self.learning_rate * db
        return self
    
    
    def predict(self, X):
        return X.dot(self.w) + self.b
```

```python
model = Linear_Regression(learning_rate = 0.001 ,no_of_iterations= 1000)
w , b = model.fit(x1,y1)
y = w[0]*x1 + w[1]*x1 + w[2]*x1 + b
plt.scatter(x1[:, 0],y1)
plt.plot(x1,y,c = '#4CAF50')
```
