## Loss/Cost functions in regression problems


**1- Mean Squared Error Cost Function general  Equation:**

The MSE loss function penalizes the model for making large errors by squaring them and this property makes the MSE cost function less robust to outliers, but it is commonly used as it's easier in differentation.

costF.PNG

**2- Mean Absolute Error Function Equation**
The corresponding cost function is the mean of these absolute errors (MAE). The MAE loss function is more robust to outliers compared to the MSE loss function.
